```
laptopspecs = spark.read.format("csv") \
    .option("header", "true") \
    .option("delimiter", "|") \
    .option("quote", "\"") \
    .option("escape", "\"") \
    .option("mode", "PERMISSIVE") \
    .option("multiline", "true") \
    .option("ignoreTrailingWhiteSpace", "true") \
    .option("ignoreLeadingWhiteSpace", "true") \
    .load("/mnt/laptopspec/raw-data/laptopspecs.csv")
```

```
laptopspecs = spark.read.format("csv").option("header","true").option("inferSchema","true").load("/mnt/laptopspec/raw-data/laptopspecs.csv")
```

```
display(laptopspecs)
```

```
laptopspecs.printSchema()
```

```
#transforming string to integer properties
laptopspecs = laptopspecs.withColumn("price",col("price").cast(IntegerType()))\
    .withColumn("rating",col("rating").cast(IntegerType()))\
    .withColumn("Core",col("Core").cast(IntegerType()))\
    .withColumn("threads",col("threads").cast(IntegerType()))\
    .withColumn("display_inches",col("display_inches").cast(IntegerType()))\
    .withColumn("ram_storage",col("ram_storage").cast(IntegerType()))\
    .withColumn("operating_system",col("operating_system").cast(IntegerType()))
```

```
laptopspecs.printSchema()
```

```
#Find the laptop with the highest price and the processor brand
laptop_highest_price_pro = laptopspecs.orderBy("price", ascending=False).select("brand_name","price","Processor_brand").show()
```
